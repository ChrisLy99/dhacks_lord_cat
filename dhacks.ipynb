#### Plotting geographical locations of BART stations and hotspots

```python
b_pair = b_a[b_a['Origin Movement ID'].isin(b_ID)].groupby('Origin Movement ID')['Origin Display Name'].unique()
```

```python
h_pair = h_a[h_a['Origin Movement ID'].isin(h_ID)].groupby('Origin Movement ID')['Origin Display Name'].unique()
```

```python
places = b_pair.append(h_pair).to_frame().reset_index()
```

```python
coords = pd.DataFrame([[3396, 37.802, -122.448],
             [3792, 37.778, -122.389],
             [3394, 37.808, -122.417],
             [3603, 37.792, -122.397],
             [3692, 37.789, -122.401],
             [3760, 37.784, -122.407]],
            columns=['ID', 'latitude', 'longitude'])
```

```python
place_coords = places.merge(coords, left_on='Origin Movement ID', right_on='ID').drop('ID', axis=1)
place_coords
```

```python
locations = place_coords[['latitude', 'longitude']]
locationlist = locations.values.tolist()
```

```python
map = folium.Map(location=[37.784, -122.407], zoom_start=13)
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup=place_coords['Origin Display Name'][point]).add_to(map)
map
```
